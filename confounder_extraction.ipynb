```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot
```

```python
# data loading
metadata=pd.read_excel('../wild_biome_r4_metadata.xlsx')
sampleData=pd.read_excel('UNPA-02-24ML+ MIXED FECES DATA TABLES_animals.XLSX',sheet_name='Sample Meta Data')
```

```python
sampleData=sampleData[20:]
sampleData
```

```python
sampleData.loc[:52,'Batch']=0
sampleData.loc[52:,'Batch']=1
```

```python
for index,row in sampleData.iterrows():
    ZA=row['CLIENT_SAMPLE_ID'].split()
    ZA='_'.join(ZA)
    sampleData.loc[index,'CLIENT_SAMPLE_ID']=ZA
```

```python
data = pd.merge(metadata, sampleData, left_on='ZA_number', right_on='CLIENT_SAMPLE_ID', how='right')
```

```python
data.to_csv('merged_metadata.csv')
```

```python
columns=pd.DataFrame(data.columns)
```

```python
columns.to_csv('Metadata_columns.csv',index=None,header=None)
```

# extraction

```python
data_confounder=data[['Zoo/Lab','Animal Species','Susceptibility (Conservative)',
                      'Susceptibility (Broad)','Animal class/order','Sex','Batch']]
```

```python
data_confounder
```

```python
data_confounder['Susceptibility (Broad)'].value_counts()
```

```python
susceptible_categories=['extremely_high_susceptibility','very_high_susceptibility','intermediate_susceptibility','high_susceptibility']
data_confounder['Susceptibility (binary)'] = data_confounder["Susceptibility (Broad)"].apply(lambda x: 1 if x in susceptible_categories else 0)
```

```python
data_confounder=data_confounder.set_index(sampleData['CLIENT_SAMPLE_ID'])
data_confounder.to_csv('confounder_info.csv')
```

```python
from scipy.stats import chi2_contingency


last_col = data_confounder.columns[-1]  
tables=[]

chi_square_results = {}

for col in data_confounder.columns[:-1]:  
    if (col != 'Susceptibility (Broad)') and (col != 'Susceptibility (Conservative)'):
        contingency_table = pd.crosstab(data_confounder[col], data_confounder[last_col])  
        tables.append(contingency_table)
        chi2, p, dof, expected = chi2_contingency(contingency_table) 
        chi_square_results[col] = {'Chi2': chi2, 'p-value': p, 'DoF': dof}

from statsmodels.stats.multitest import multipletests


chi_square_data_confounder = pd.DataFrame.from_dict(chi_square_results, orient='index')

p_values = chi_square_data_confounder["p-value"].values  

_, p_bh, _, _ = multipletests(p_values, method='fdr_bh')


chi_square_data_confounder["Benjamini-Hochberg p-value"] = p_bh

print(chi_square_data_confounder.sort_values(by="p-value"))
```

```python
chi_square_data_confounder.sort_values(by="p-value").to_csv('confounder.csv')
```
